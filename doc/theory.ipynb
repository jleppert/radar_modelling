# Fundamentals of radar signal processing for Stepped Frequency Continuous Wave Radar

## Simple Radar Model
Figure x depicts a block diagram of a radar observing a small "point" target at some distance $R$ from the antenna. The system consists of a transmitter, circulator, single antenna and receiver chain

### Transmit Signal
The transmitter transmits a burst of RF energy modelled by the signal $v_{TX}(t)$, which can be thought of as the voltage waveform on the terminals of the antenna.

### Response from a single point target

Assuming that the signal bandwidth is less than that of the receiving antenna, we can model the received signal as a delayed, scaled version of the transmitted pulse

\begin{align}
v_{RX} = a_1 v_{TX}(t- \tau)
\end{align}

where $\tau = \frac{2R}{c}$ is the time delay and the constant $a_1$ is a factor which accounts for the various system and propagation effects. For a narrowband radar, this factor can be related to the parameters of the standard radar equation by

\begin{align}
a_1 \propto \sqrt{\frac{G^2\sigma\lambda^2} {(4 \pi)^3 R^4} }
\end{align}

where $G$ is the antenna gain, $\sigma$ is the radar cross section of the target and $\lambda$ is the wavelength (all parameters regarded as constant over the narrow fractional bandwidth, say $1$%.

Consider $N$ point targets $R_i$. As the the complete system is linear, the received signal is simply the sum of the response, i.e.
$$ v_{RX}(t) = \sum_{i=1}^{N} a_{i} v_{TX}(t-\tau_i)$$
where $a_i$ and $\tau_i$ model the amplitudes and the time delays of the targets. For scenes containing more complicated backscatter mechanisms such as extended targets including multipath effects, the interaction of the electromagnetic wave with a stationary scene (as viewed from the location of the radar) can be completely described by the impulse response $\eta(t)$. One can think of the response from the scene in terms of the linear model depicted in Figure below




If the radar transmits some arbitrary waveform  $v_{TX}(t)$, the received signal is then the convolution of $\eta(t)$$ and $v_{TX}$, i.e. 

\begin{align}
v_{RX}(t)  = \eta(t) \circledast  v_{TX}(t) = \int_{-\infty}^{\infty} \eta(t) v_{TX}(t-\tau)dr
\end{align}

For the special case where the scene may be modelled as a collection of point targets, the impulse response of the scene is the weighted sum 

\begin{align}
\eta(t) = \sum_{i=1}^{N}a_i \delta(t-\tau_i)
\end{align}

where $a_i$ is the weight of the echo from the i^{th} target, and $\tau_i = \frac{2R_i}{c}$ is the delay. The corresponding received signal is, by convolution,

\begin{align}
v_{RX}(t) &= \eta(t) \circledast v_{TX}(t) \\ 
&= \left[ \sum_{i=1}^{N}a_i \delta(t-\tau_i)  \right] \circledast v_{TX}(t) \\
&= \sum_{i=1}^{N} \left[ a_i \delta(t-\tau_i) \circledast v_{TX}(t) \right] \\
&= \sum_{i=1}^{N} a_i v_{TX}(t-\tau_i)
\end{align}

as expected.
